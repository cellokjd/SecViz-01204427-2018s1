```python
day_1 = {}
day_2 = {}
for i in range(0,24):
    col.append(i+0.00)
    col.append(i+0.15)
    col.append(i+0.30)
    col.append(i+0.45)
    day_1[str(i+0.00)] = 0 
    day_1[str(i+0.15)] = 0 
    day_1[str(i+0.30)] = 0
    day_1[str(i+0.45)] = 0
    day_2[str(i+0.00)] = 0
    day_2[str(i+0.15)] = 0 
    day_2[str(i+0.30)] = 0 
    day_2[str(i+0.45)] = 0 
for index, row in df.iterrows():
    login_time = (datetime.utcfromtimestamp(row['login']/1000))
    if(row['logout']!='-'):
        logout_time = (datetime.utcfromtimestamp(int(row['logout'])/1000))
#         print(login_time.strftime('Date %d Time %H:%M')+' TO ' + logout_time.strftime('Date %d %H:%M') )
        if(login_time.strftime('%d') == '01' and logout_time.strftime('%d') == '01'):
            temp = int(login_time.strftime('%H'))+int(login_time.strftime('%M'))/100
            temp2 = int(logout_time.strftime('%H'))+int(logout_time.strftime('%M'))/100
            for i in range(0,24):
                if(temp <= i and i <= temp2):
                    day_1[str(i+0.00)] += 1
                if(temp <= (i+0.15) and (i+0.15) <= temp2):
                    day_1[str(i+0.15)] += 1
                if(temp <= (i+0.30) and (i+0.30) <= temp2):
                    day_1[str(i+0.30)] += 1
                if(temp <= (i+0.45) and (i+0.45) <= temp2):
                    day_1[str(i+0.45)] += 1
            print(login_time.strftime('%d-%m-%y'))
        elif(login_time.strftime('%d') == '01' and logout_time.strftime('%d') == '02'):
            temp = int(login_time.strftime('%H'))+int(login_time.strftime('%M'))/100
            temp2 = int(logout_time.strftime('%H'))+int(logout_time.strftime('%M'))/100
            for i in range(0,24):
                if(i < 22):
                    if(i <= temp2):
                        day_2[str(i+0.00)] += 1
                    if((i+0.15) <= temp2):
                        day_2[str(i+0.15)] += 1
                    if((i+0.30) <= temp2):
                        day_2[str(i+0.30)] += 1
                    if((i+0.45) <= temp2):
                        day_2[str(i+0.45)] += 1
                else:
                    if(temp >= i ):
                        day_1[str(i+0.00)] += 1
                    if(temp >= (i+0.15)):
                        day_1[str(i+0.15)] += 1
                    if(temp >= (i+0.30)):
                        day_1[str(i+0.30)] += 1
                    if(temp >= (i+0.45) ):
                        day_1[str(i+0.45)] += 1 
            print(login_time.strftime('%d-%m-%y %H-%M'))
        else:
            temp = int(login_time.strftime('%H'))+int(login_time.strftime('%M'))/100
            temp2 = int(logout_time.strftime('%H'))+int(logout_time.strftime('%M'))/100
            for i in range(0,24):
                if(temp <= i and i <= temp2):
                    day_2[str(i+0.00)] += 1
                if(temp <= (i+0.15) and (i+0.15) <= temp2):
                    day_2[str(i+0.15)] += 1
                if(temp <= (i+0.30) and (i+0.30) <= temp2):
                    day_2[str(i+0.30)] += 1
                if(temp <= (i+0.45) and (i+0.45) <= temp2):
                    day_2[str(i+0.45)] += 1
            
    else:
#         print(login_time.strftime('Date %d Time %H:%M') + " UP ")
        temp=1
```
